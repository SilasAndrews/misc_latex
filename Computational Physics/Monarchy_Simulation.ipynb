```python
import numpy as np
from matplotlib import pyplot as plt
import random as rand
```

```python
N=1000 #number of players to simulate
tmax = 500 #number of turns to simulate
bins = np.arange(0,41)
loc = []
for t in range(0,tmax):
    di1 = np.random.randint(1,high=7, size=N)
    di2 = np.random.randint(1,high=7, size=N)
    loc.append((di1+di2))
    if t>0:
        loc[t] += loc[t-1]
    for i in range(0,N):
        loc[t][i] = loc[t][i]%40
        if loc[t][i] == 30:
            loc[t][i] = 10
```

```python
plot=499 #which turn do you want to show the distribution of (add 1 for actual turn number)
print('Average position after turn',plot+1,':',np.average(loc[plot]))
plt.hist(loc[plot],bins=bins)
plt.show()
```

```python
num = 50 #number of plots (turns) to show, higher than 20 can take a few minutes
fig,ax = plt.subplots(num,1,figsize=(10,5*num))
for i in range(0,num):
    ax[i].hist(loc[i],bins=bins)
    ax[i].set_title("turn " + str(i+1))
```

```python
print('Average after n turns')
for i in range(0,100):
    print(i+1,'turns: ',np.average(loc[i]))
```
