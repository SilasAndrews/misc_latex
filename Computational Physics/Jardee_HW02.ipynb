```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
```

a. A function to describe the distance between Earth and Mars. This is using the constants given in class and the law of  cosines.

```python
def mars_dis(d):
    """
    Find the Distance, in kilometers, that mars appears to be from Earth on some day 'd' from Jan 1, 2000

    Input
    ----------
    d: This is the day that I want to evaluate things at

    Output
    ----------
    Distance between Earth and Mars for the given input day 'd' in km
    """

    Pe = 365.256    #Period of the Earths orbit (days)
    Pm = 686.980    #Period of Mars orbit (days)
    d0 = 187        #Day in the year 2000 that mars crosses theta = 0
    Re = 1.496e8    #Radius of Earths Orbit (km)
    Rm = 2.279e8    #Radius of Mars orbit (km)

    phi_e = 2*np.pi*d/Pe
    phi_m = 2*np.pi*(d-d0)/Pm
    
    #Uses the law of cosines
    return m.sqrt(Re**2 + Rm**2 - 2*Re*Rm*np.cos(phi_m-phi_e)) 
```

```python
# if you don't include the 27th, June 27, 2000 -> 178 days from Jan 1, 2000
print(str.format("On June 27, 2000 the planets are {:.3e}km apart.\n",mars_dis(178)))

# if you don't include the 1st, Sept 1, 2020 -> 7549 days from Jan 1, 2000
print(str.format("On Sept 1, 2020 the planets are {:.3e}km apart.\n",mars_dis(7549)))

#Days given by https://www.timeanddate.com/date/durationresult.html
```

c. Plotting the Distance as a function of Time. One notable part of the code is that I had to vecotize the function to get two vectors to plot against each other. I think that if I used a np array instead of a linespace the plot would work with the funciton, however this solution works and is more straight forward I think. 

```python
DAY = 7305 #Number of days between Jan 1, 2000 and Jan 1, 2020

x = np.linspace(0,7305,7300) #Goes from 0 to 7305, and divides it into 7300 equal pieces.
mars_dis_vects= np.vectorize(mars_dis)
#Line of code via https://mindmajix.com/community/49677/only-size--arrays-can-be-converted-to-python-scalars#:~:text=0-,The%20error%20called%20only%20length%2D1%20arrays%20can%20be%20converted,the%20function%3A%20f2%20%3D%20np.
#Allows for pyplot to recognize mars_dis as a vector since the function doesn't take in vectors

plt.figure(figsize = (17,17))
plt.plot(x, mars_dis_vects(x)) 
plt.title('Distance from Earth to Mars', fontsize=30)
plt.ylabel('Distance (km)',size = 20)
plt.xlabel('Days From Jan 1, 2000', size = 20)
plt.show()
```
